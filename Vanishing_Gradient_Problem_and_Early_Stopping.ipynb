**Early Stopping**

```python
x,y= make_moons(n_samples= 100, noise=0.1, random_state= 101)
```

```python
plt.scatter(x[:,0], x[:,1], c=y, s=100)
plt.show()
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, random_state= 125)
```

```python
demo= Sequential()
demo.add(Dense(250, input_dim=2, activation= 'relu'))
demo.add(Dense(1, activation= 'sigmoid'))
demo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history= demo.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3000, verbose=0)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
callback= EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=15,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
)
```

```python
history= demo.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=3000, callbacks= callback)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
